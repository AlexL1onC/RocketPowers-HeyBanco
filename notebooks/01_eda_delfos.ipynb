```python
import torch
from transformers import pipeline

# 1. Configurar el pipeline generativo
# Usamos el modelo destilado de 1.5B parámetros (ligero y potente)
model_id = "deepseek-ai/DeepSeek-R1-Distill-Qwen-1.5B"

print("Cargando a Delfos (DeepSeek-R1)...")
delfos_pipe = pipeline(
    "text-generation",
    model=model_id,
    torch_dtype=torch.float16, # Para que consuma menos memoria
    device_map="auto"           # Detecta GPU o CPU automáticamente
)

def probar_delfos(indice_fila, dataframe):
    mensaje_usuario = dataframe.iloc[indice_fila]['input']
    respuesta_original = dataframe.iloc[indice_fila]['output']
    
    # Prompt de Sistema para darle la personalidad de Delfos
    prompt = f"""<|im_start|>system
Eres Delfos, el asesor financiero inteligente de HeyBanco. 
Tu lenguaje es el español. Eres empático, técnico y estratégico.
Basate en los datos del cliente para dar una respuesta útil.
<|im_end|>
<|im_start|>user
Mensaje del cliente: {mensaje_usuario}
<|im_end|>
<|im_start|>assistant
<|thought|>
Analizaré la duda sobre '{mensaje_usuario}' y responderé de forma proactiva.
<|thought|>"""

    # Generación
    outputs = delfos_pipe(
        prompt, 
        max_new_tokens=150, 
        temperature=0.3, # Bajo para evitar alucinaciones
        do_sample=True,
        pad_token_id=50256 # Valor estándar para evitar warnings
    )
    
    respuesta_delfos = outputs[0]["generated_text"].split("<|im_end|>\n<|im_start|>assistant\n")[-1].strip()
    
    # Visualización comparativa
    print(f"\n{'='*50}")
    print(f"CASO EN PARQUET (Fila {indice_fila})")
    print(f"{'='*50}")
    print(f"USUARIO: {mensaje_usuario}")
    print(f"-"*30)
    print(f"HAVI v1 (Original): {respuesta_original}")
    print(f"-"*30)
    print(f"DELFOS (DeepSeek): {respuesta_final_limpia(respuesta_delfos)}")

def respuesta_final_limpia(texto):
    # DeepSeek suele incluir el proceso de pensamiento entre etiquetas <|thought|>
    # Aquí limpiamos eso para la respuesta final al cliente
    if "</|thought|>" in texto:
        return texto.split("</|thought|>")[-1].strip()
    return texto

# --- EJECUTAR PRUEBA ---
# Prueba con la fila 0 que encontraste antes
probar_delfos(0, df)
```

```python
import torch
import gc
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline

# 1. Limpieza agresiva de memoria
gc.collect()
if torch.backends.mps.is_available():
    torch.mps.empty_cache()

model_id = "deepseek-ai/DeepSeek-R1-Distill-Qwen-1.5B"

# 2. Cargar tokenizador
tokenizer = AutoTokenizer.from_pretrained(model_id)

# 3. Cargar modelo optimizado para Apple Silicon
print("Cargando Delfos en la GPU de tu Mac (MPS)...")
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    torch_dtype=torch.float16, # Mitad de precisión = mitad de RAM
    low_cpu_mem_usage=True,
    device_map="mps"           # <--- Forzamos el uso de Metal (GPU de Mac)
)

delfos_pipe = pipeline("text-generation", model=model, tokenizer=tokenizer)
print(f"✅ Delfos está viviendo en: {model.device}")
```
