```python
import duckdb

# Conectarte directo al archivo, sin pasar por FastAPI
con = duckdb.connect("havi.duckdb")

print("--- Tablas Disponibles ---")
print(con.execute("SHOW TABLES").df())

print("\n--- Estructura de la tabla 'conversaciones' ---")
# Esto te mostrará las columnas y el tipo de dato
print(con.execute("DESCRIBE conversaciones").df())

print("\n--- Muestra de 3 registros ---")
print(con.execute("SELECT * FROM conversaciones LIMIT 3").df())

print("\n--- Promedio de consultas por usuario ---")
print(con.execute("SELECT AVG(consultas) AS promedio_consultas FROM (SELECT user_id, COUNT(*) AS consultas FROM conversaciones GROUP BY user_id)").df())

con.close()
```
